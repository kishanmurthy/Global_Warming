# Insight on Global Warming

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("GlobalLandTemperaturesByMajorCity.csv")
```

```python
data.dropna(axis=0 , inplace=True)
```

```python
data["Month"]=data["dt"].map(lambda x:int(x.split("-")[1]))
```

```python
data["Year"]=data["dt"].map(lambda x:int(x.split("-")[0]))
```

```python
data.drop(axis=1 ,labels="dt" ,inplace=True)
```

```python
data.drop(axis=1 ,labels="AverageTemperatureUncertainty" ,inplace=True)
```

```python
bangalore_avg=data[data["City"]=="Bangalore"].groupby("Year").mean()["AverageTemperature"]
```

```python
plt.plot(bangalore_avg)
```

```python
def group_a_month(city,mon):
    tmp=data[(data["City"]==city) & (data["Month"]==mon)][["Year" , "AverageTemperature"]]
    tmp.set_index("Year" ,inplace="True")
    return tmp
```

```python
def group_all_months(city):
    month=[]
    month.append(0);
    for i in range(1,13):
        month.append(group_a_month(city,i))
    return month    
    
        
    
```

```python
def group_all_cities():
    cities={}
    for city in data["City"].unique():
        cities[city]=group_all_months(city)
        
    return cities
```

```python
plt.plot(group_all_months("Bangalore")[12])
```

```python
def plot_a_city(city):
    fig,axes =plt.subplots(nrows=12 ,ncols=1,figsize=(13,70))
    for i in range(0,12):
        axes[i].plot(city[i+1])
        axes[i].set_xlabel("Years")
        axes[i].set_ylabel("Temperature")
        axes[i].set_title("Month " + str(i+1))
        
    
```

```python
data_set=group_all_cities()
```

```python
plot_a_city(data_set["Chicago"])
```

```python

def find_average_temp_city(city):
    years=data[data["City"]==city].groupby("Year").mean()
    year=years.index.values
    for i in year:
         if years.loc[i]["Month"]!=6.5:
            years.drop(i,axis=0 ,inplace=True)
    years.drop("Month",axis=1,inplace=True)
    return years
            
```

```python
def group_all_cities_avg():
    cities={}
    for city in data["City"].unique():
        cities[city]=find_average_temp_city(city)
        
    return cities
    
```

```python
data_set_avg=group_all_cities_avg()
```

```python
a=find_average_temp_city("Bangalore")
```

```python
miss_years=[]
for year in range(1750,2016):
    for city in data_set_avg:
        try:
            data_set_avg[city].loc[year]
        except:
            if year not in miss_years:
                miss_years.append(year)
                break
            
```

```python
def find_global_avg():
    avg=pd.DataFrame(columns=['Year','AverageTemperature'])
    avg["Year"]=list(range(1899,2013))
    avg.set_index("Year",inplace=True)
    for year in range(1899,2013):
        mysum=0;
        for city in data_set_avg:
            mysum=mysum+data_set_avg[city].loc[year]
        avg.loc[year]=mysum/100
    return avg
```

```python
cities_ml={}
for city in data_set_avg:
    tmp=data_set_avg[city].reset_index()
    tmp.set_index("Year", inplace=True)
    for i in range(1750,1899):
        try:
            tmp.drop(i,axis=0,inplace=True)
        except:
            pass
    tmp.reset_index(inplace=True)
    x=tmp[["Year"]]
    y=tmp[["AverageTemperature"]]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
    t=LinearRegression()
    t.fit(x_train,y_train)
    t.coef_[0][0]
    cities_ml[city]=[[x,y,x_train, x_test, y_train, y_test], t ,t.coef_[0][0]]
    
```

```python
def display_save(city):
    tmp=data_set_avg[city].reset_index()
    tmp.set_index("Year", inplace=True)
    for i in range(1750,1899):
        try:
            tmp.drop(i,axis=0,inplace=True)
        except:
            pass
    fig,axes=plt.subplots()
    axes.plot(tmp)
    axes.set_xlabel("Year")
    axes.set_ylabel("AverageTemperature")
    axes.set_title(city)
    fig.savefig(city + ".png" ,dpi=200)
```

```python
display_save("Moscow")
```

```python
display_save("Chongqing")
```

```python
def display_save_reg(city):
    tmp=data_set_avg[city].reset_index()
    tmp.set_index("Year", inplace=True)
    for i in range(1750,1899):
        try:
            tmp.drop(i,axis=0,inplace=True)
        except:
            pass
    tmp.reset_index(inplace=True)
    sns_plot=sns.jointplot(x='Year',y='AverageTemperature',data=tmp,kind='reg')
    sns_plot.savefig(city + "_reg.png" ,dpi=200)
```

```python
display_save_reg("Moscow")
```
